```python
#methods-  get values
profile={'name':'nitika','age':20,'branch' : 'Ai&Ml'}
age = profile.get('age')
print(age)
age = profile.get('age_2','Not found')
print(age)
```

```python
#keys ,values and items method #pop method
profile={'name':'nitika','age':20,'branch' : 'Ai&Ml'}
keys =profile.keys()
print(keys)
values =profile.values()
print(values)
all_items = profile.items()
print(all_items)
popped = profile.pop('age')
print(popped)
print(profile)
popped_i = profile.popitem()
print(popped_i)
print(profile)
cleared = profile.clear()
print(profile)
```
