```python
import pandas as pd
```

```python
# Load dataset
dataset = pd.read_excel("/content/drive/MyDrive/B.Tech. AIML- VI/B.Tech. AIML Section C/Datasets/Influenza_surveillance_Data.xlsx")
```

```python
dataset['Pathogen Test Result'].value_counts()
```

```python
# Separate features and target
X = dataset.drop("Pathogen Test Result", axis=1)#Features
y = dataset["Pathogen Test Result"]#Target
```

```python
X.shape
```

```python
y.shape
```

```python
# Import required libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
# Build the neural network model
model = Sequential()

# Hidden Layer 1
model.add(Dense(
    units=3,
    activation='relu',
    input_shape=(34,)
))
# Hidden Layer 2
model.add(Dense(
    units=2,
    activation='relu'
))

# Output Layer
model.add(Dense(
    units=1,
    activation='sigmoid'
))
```

```python
# Compile the model

model.compile(
    optimizer=Adam(learning_rate=0.01),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
# Train the model

history = model.fit(
    X,
    y,
    epochs=50,
    batch_size=4,
    verbose=1
)
```

```python
# Evaluate the model

loss, accuracy = model.evaluate(X, y)

print("Loss:", loss)
print("Accuracy:", accuracy)
```

```python
# Select a single patient
new_patient = X.iloc[[0]]

prediction = model.predict(new_patient)

print("Predicted Probability:", prediction)

if prediction >= 0.5:
    print("Prediction: Influenza Positive")
else:
    print("Prediction: Influenza Negative")
```
