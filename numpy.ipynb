```python
# reshaping & manipulating arrays
"""
reshaping = changing the dimension without changing the array
1d -2d,2d - 3d ,it does not create copy ,it ceate a view and it affect our array
reshape(rows,columns)specify new shape if dimension match
arr.reshape()
"""
arr = np.array([1,2,3,4,5,6])
reshape_arr = arr.reshape(3,2)
print(reshape_arr)
```

```python
#flattering array -change multidimensional to 1d array
# .ravel() - view
# .flatten() - copy'
import numpy as np
arr_2d = np.array([[1,2,3],[4,5,6]])
print(arr_2d.ravel())
print(arr_2d.flatten())
```

```python
"""
np.insert(array,index,value,axis = None)
axis is none thrn it is flatten - 1d array
axis is 0 ,1 then it is in 2d array
axis - 0 then row wise
axis -1 then column wise insert data
"""
arr = np.array([10,20,30,40,50])
new_arr = np.insert(arr,3,70,axis = None)
print(new_arr)
arr_2d = np.array([[10,20],[30,40]])
new_arr2d = np.insert(arr_2d,2,70,axis =1)
print(new_arr2d)
new_arr2c = np.insert(arr_2d,1,70,axis =0)
print(new_arr2c)
```

```python
# append - add the element in the end
arr = np.array([10,20,30,40,50])
new_arr = np.append(arr,[40,50,60])
print(new_arr)
```

```python
# concatenation - join different array
#np.concatenate((array1,array2),axis = 0,1)
arr1 = np.array([[10,20,30],[40,50,60]])
print(arr1)
arr2 = np.array([[50,60,70],[80,90,100]])
print(arr2)
new_arr = np.concatenate((arr1,arr2),axis =0)
print(new_arr)
```

```python
#Stacking - combining vertically and horizentally
# vstack() - row wise
# hstack() - column wise
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print("row wise")
print(np.vstack((arr1,arr2)))
print("column wise ")
print(np.hstack((arr1,arr2)))
```

```python
# using loop
prices =[100,200,300]
discount = 10  # 10% discount 
final_prices = []
for price in prices:
    final_price = price -(price*discount/100)
    final_prices.append(final_price)
print(final_prices)
```

```python
# broadcasting and vectorization
"""
broadcasting - applying multiple mathematical opertions without using loop
"""
import numpy as np
prices = np.array([100,200,300])
discount = 10
final_prices = prices -(prices *discount/100)
print(final_prices)
```

```python
#handling missing values in numpy
# isnan function(array)
import numpy as np
arr = np.array([1,2,np.nan,4,5,np.nan,6,7])
print(np.isnan(arr))
# where the output is true value is missing
```

```python
# nan to num - np.nan_to_num(array,nan-value ,default =0)
arr = np.array([2,3,4,5,6,np.nan,7,np.nan])
print(np.nan_to_num(arr,nan = 100))
print(np.nan_to_num(arr))
```

```python
# to handle infinite values
#np.isinf()
# posinf = positive infinity
# neginf = negative infinity
import numpy as np
arr = np.array([1,2,np.inf,4,-np.inf,6])
print(np.isinf(arr))
cleaned_arr = np.nan_to_num(arr,posinf=1000,neginf=1000)
print(cleaned_arr) 
```
